```python
def plot_greedy_modularity_communities(g):
    # Definir layout (posicionamento dos ns)
    pos = nx.spring_layout(remove, k=0.7)

    # Detectar as comunidades no g (usando um algoritmo de deteco de comunidades)
    comunidades = nx.community.greedy_modularity_communities(g)

    # Criar um dicionrio que mapeia cada n a sua comunidade
    comunidade_dict = {}
    for i, comunidade in enumerate(comunidades):
        for node in comunidade:
            comunidade_dict[node] = i

    print(len(comunidades))
    plt.figure(figsize=(10,10))

    cores = ['red', 'blue', 'green', 'yellow', 'purple', 'pink', 'brown', 'orange', 'white', 'black', 'teal']  
    nx.draw_networkx_nodes(g, pos, nodelist=comunidade_dict.keys(),
                        node_color=[cores[comunidade_dict[node]] for node in comunidade_dict],
                        node_size=10, alpha=0.5)

    nx.draw_networkx_edges(g, pos, alpha=0.1)

    plt.axis('off')

    return plt.show()
```

```python
plot_greedy_modularity_communities(remove)
```
